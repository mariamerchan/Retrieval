```
def renombras_dataset(path_DATASET,path_DATASET_ORI):

  print("¿Cual es el número entero de la ultima carpeta?")
  actual_numero_carpeta = input()

  print(f"La carpeta actual, {actual_numero_carpeta}")

  #Convertir el numero a entero
  actual_numero_carpeta= int(actual_numero_carpeta)

  path_original = path_DATASET#1
  path_destino=path_DATASET_ORI #2

  cont_carpetas=0
  cont_imagenes=0
  #actual_numero_carpeta=0

  #For que recorre las carpetas(sujetos) del path_original
  for cj in os.listdir(path_original):
    cont_carpetas=cont_carpetas+1
    actual_numero_carpeta = actual_numero_carpeta+1
    num_imagen=1
    #For que recorre la imagen de cada carpeta(sujeto)
    for nj in os.listdir(path_original + '/' + cj):
      cont_imagenes=cont_imagenes+1

      #-------------
      #cj   -> Carpeta del sujeto 
      #nj   -> Nombre de la imagen del sujeto
      #path_original -> Ruta de la imagen
      imgPath = path_original + '/' + cj + '/' + nj
      #-------------
    
      #Abrir Imagen (Lee la imagen en formtado BGR)
      imgBGR = cv2.imread(imgPath)   

      #Nombre y extension
      filename, ext = os.path.splitext(os.path.basename(imgPath))

      nom_carpeta=str(actual_numero_carpeta)+'_'+cj
      nom_carpeta=nom_carpeta.replace('-','_')
      nom_carpeta=nom_carpeta.replace(' ','')

      #Guarda imagenes imagenes con cuadros
      os.makedirs(path_destino+ '/'+nom_carpeta, exist_ok=True)
      #filename=(path_destino+ '/' + cj + '/' + nj) 
      RutaGuardar=(path_destino+ '/' + nom_carpeta + '/' + nom_carpeta + '_'+ str(num_imagen) + ext) 
      num_imagen=num_imagen+1
      cv2.imwrite(RutaGuardar, imgBGR) 

  print("TERMINO !! ;)")
  print("Cantidad de carpetas procesadas", cont_carpetas)
  print("Cantidad de imagenes procesadas", cont_imagenes)


```

```
def caso_opcion(opcion):
  if opcion == 1:
    print('Perros')
    path_DATASET = '3.Datasets_original/1.DATASET_PERROS' #1
    path_DATASET_ORI = '3.Datasets/1.DATASET_PERROS' #2
    renombras_dataset(path_DATASET,path_DATASET_ORI)

    print(os.path.isdir(path_DATASET))
  elif opcion == 2:
    print('Buscar Perro')
    path_DATASET = '3.Datasets_original/3.DATASET_ENTRADA_PERROS' #1
    path_DATASET_ORI = '3.Datasets/3.DATASET_ENTRADA_PERROS' #2
    renombras_dataset(path_DATASET,path_DATASET_ORI)
    print(os.path.isdir(path_DATASET))
  elif opcion == 3:
    print('Gatos')
    path_DATASET = '3.Datasets_original/2.DATASET_GATOS' #1
    path_DATASET_ORI = '3.Datasets/2.DATASET_GATOS' #2
    renombras_dataset(path_DATASET,path_DATASET_ORI)
    
    print(os.path.isdir(path_DATASET))
  elif opcion == 4:
    print('Buscar Gato')
    path_DATASET = '3.Datasets_original/4.DATASET_ENTRADA_GATOS' #1
    path_DATASET_ORI = '3.Datasets/4.DATASET_ENTRADA_GATOS' #2
    renombras_dataset(path_DATASET,path_DATASET_ORI)
    print(os.path.isdir(path_DATASET))
  else:
    #seleccion_opcion()
    print('xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx')
    print('Error: La opción ingresada no esta dentro de las opciones, por favor vuelva a ingresar la opción')
    print('xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx')
    print('')
    print('')
    print('')
    seleccion_opcion()

seleccion_opcion()
```
