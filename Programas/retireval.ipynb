#**3. Cargamos el archivo de características Ejemplo**

```python
dataset = pt.get_dataset("vaswani")
print(type(dataset))
bm25 = pt.BatchRetrieve(dataset.get_index(), wmodel="BM25")
dph = pt.BatchRetrieve(dataset.get_index(), wmodel="DPH")
pt.Experiment(
    [bm25, dph], 
    dataset.get_topics(),
    dataset.get_qrels(),
    eval_metrics=["map", "P_10", "P_20", "ndcg_cut_20"]
)
```
