```
import numpy as np
import pandas as pd

def funcion_lectura_carpetas(DATOS_UNO,CARPETA_act, MEDIDA,CARPETA): #, MEDIDA PENDIENTE

  print('*******************')
  print('DATOS_UNO : '+DATOS_UNO)
  print(os.path.isfile(DATOS_UNO))
  print('DATOS_DOS : '+CARPETA_act)
  
  print('MEDIDA : '+ str(MEDIDA))
  print('CARPETA : '+ CARPETA)
  print('*******************')

  carpetas_ruta=CARPETA_act.split(sep='/')
  carpeta_k=carpetas_ruta[2]
  

  if(os.path.isfile(DATOS_UNO)):
    #Imagen de Datasets
    conteo_x_imagen = pd.read_csv(DATOS_UNO,header=1 ,dtype=str)

    #CREAR ARCHIVO
    ruta_almacenamiento="8.Medidas_Desempeño/X_"+CARPETA
    os.makedirs(ruta_almacenamiento, exist_ok=True)
    ruta_almacenamiento=ruta_almacenamiento+"/"+carpeta_k+".csv"
    archivo_desempenyo= open(ruta_almacenamiento, "w+") 

    archivo_desempenyo.write("Sujeto")
    archivo_desempenyo.write(",")
    archivo_desempenyo.write("# Img Dataset")
    archivo_desempenyo.write(",")
    archivo_desempenyo.write("# Img Recup-K1")
    archivo_desempenyo.write(",")
    archivo_desempenyo.write("# Img Recup-K2")
    archivo_desempenyo.write(",")
    archivo_desempenyo.write("# Img Recup-K3")
    archivo_desempenyo.write(",")
    archivo_desempenyo.write("# Img Recup-TK3")
    archivo_desempenyo.write("\n")

    for archivo_medida in os.listdir(CARPETA_act):
      
      print(archivo_medida)

      #Obtenemos el nombre del sujeto
      KA=archivo_medida.split(sep='__')
      sujeto_evaluado=KA[0]

      #Lectura del archivo
      lectura=CARPETA_act+'/'+archivo_medida
      archivo_x_sujeto = pd.read_csv(lectura,header=0)

      #Ordenamos de menor a mayor
      #archivo_x_sujeto_order=pd.DataFrame(np.sort(archivo_x_sujeto.values,axis=0),index=archivo_x_sujeto.index,columns=archivo_x_sujeto.columns)
      df = pd.DataFrame(archivo_x_sujeto)
      archivo_x_sujeto_order = df.sort_values(['# Valor Distancia Euclidiana'], ignore_index=True)

      #Cantidad de Imagenes en el Dataset
      Img_Ini_escribir='NO APLICA'
      sujeto_escribir=sujeto_evaluado
      for j in range(len(conteo_x_imagen.index)):
        sujeto_escribir=archivo_x_sujeto['# Sujetos Query'][1]
        sujeto_cant=conteo_x_imagen['Sujeto'][j]
        if(sujeto_escribir==sujeto_cant):
          Img_Ini_escribir=conteo_x_imagen['# Imagenes'][j]
          break;
      
      #Cantidad de Imagenes en el Recuperadas
      Img_Res_escribir_K0=0
      Img_Res_escribir_K1=0
      Img_Res_escribir_K2=0
      Img_Res_escribir_KT=0
      for i in [0,1,2]: #K=3
        sujeto_escribir=archivo_x_sujeto_order['# Sujetos Query'][i]
        sujeto_dataset=archivo_x_sujeto_order['# Sujetos Dataset'][i]
        if(sujeto_escribir==sujeto_dataset):
          if(i==0):
            Img_Res_escribir_K0=1
          elif(i==1):
            Img_Res_escribir_K1=1
          elif(i==2):
            Img_Res_escribir_K2=1

          Img_Res_escribir_KT=Img_Res_escribir_KT+1
          #print(archivo_x_sujeto_order['# Valor Distancia Euclidiana'][i])
      
      archivo_desempenyo.write(sujeto_escribir)
      archivo_desempenyo.write(",")
      archivo_desempenyo.write(str(Img_Ini_escribir))
      archivo_desempenyo.write(",")
      archivo_desempenyo.write(str(Img_Res_escribir_K0))
      archivo_desempenyo.write(",")
      archivo_desempenyo.write(str(Img_Res_escribir_K1))
      archivo_desempenyo.write(",")
      archivo_desempenyo.write(str(Img_Res_escribir_K2))
      archivo_desempenyo.write(",")
      archivo_desempenyo.write(str(Img_Res_escribir_KT))
      archivo_desempenyo.write("\n")
    archivo_desempenyo.close()
  
```
