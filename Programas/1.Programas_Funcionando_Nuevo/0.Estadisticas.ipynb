```
def contar_imagenes(RUTAS_CARPETAS,nombre_archivo):

  archivo_datos="0. Estadisticas/"+nombre_archivo+".csv"
  
  archivo_datos_escribir= open(archivo_datos, "w+") 

  archivo_datos_escribir.write(nombre_archivo)
  archivo_datos_escribir.write("\n")
  archivo_datos_escribir.write("# Ruta")
  archivo_datos_escribir.write(",")
  archivo_datos_escribir.write("# Carpetas")
  archivo_datos_escribir.write(",")
  archivo_datos_escribir.write("# Sujetos")
  archivo_datos_escribir.write(",")
  archivo_datos_escribir.write("# Imagenes")
  archivo_datos_escribir.write("\n")
  
  for ruta in RUTAS_CARPETAS:
    print("***************************************************")
    print(ruta)
    cadena = ruta
    separador = "/"
    nombre_carpeta = cadena.split(separador)
    
    #CONTADOR DE SUJETOS 
    cont_carpetas = 0
    #CONTADOR DE IMAGENES 
    num_imagen_total=0
    #For que recorre las carpetas(sujetos) de cada Ruta
    for sujeto in os.listdir(ruta):
      cont_carpetas=cont_carpetas+1
      num_imagen=0
      #For que recorre la imagen de cada carpeta(sujeto)
      for imagen in os.listdir(ruta + '/' + sujeto):
        num_imagen=num_imagen+1
      num_imagen_total=num_imagen_total+num_imagen

    archivo_datos_escribir.write(nombre_carpeta[-2])
    archivo_datos_escribir.write(",")
    archivo_datos_escribir.write(nombre_carpeta[-1])
    archivo_datos_escribir.write(",")
    archivo_datos_escribir.write(str(cont_carpetas))
    archivo_datos_escribir.write(",")
    archivo_datos_escribir.write(str(num_imagen_total))
    archivo_datos_escribir.write("\n")

    print("Cantidad de carpetas procesadas", cont_carpetas)
    print("Cantidad de imagenes procesadas", num_imagen_total)
  archivo_datos_escribir.close()  # Cerrar archivo
  print("TERMINO !! ;)")
```
