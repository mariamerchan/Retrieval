```

def preprocesamiento_imagen(path_MANUAL,path_ROSTRO_REDIMENSIONADO):
  cont_imagen=0
  #For que recorre las carpetas(sujetos) del path_DATASET
  for cj in os.listdir(path_MANUAL):
      
      #For que recorre la imagen de cada carpeta(sujeto)
      for nj in os.listdir(path_MANUAL + '/' + cj):

        cont_imagen=cont_imagen+1
        #-------------
        #cj   -> Carpeta del sujeto 
        #nj   -> Nombre de la imagen del sujeto
        #path_MANUAL -> Ruta de la imagen
        imgPath = path_MANUAL + '/' + cj + '/' + nj
        print(imgPath)
        #-------------
      
        #Abrir Imagen (Lee la imagen en formtado BGR)
        imgBGR = cv2.imread(imgPath)

        #Corta y guarda imagenes
        # Crear la carpeta si no esta creada 
        os.makedirs(path_ROSTRO_REDIMENSIONADO+ '/'+cj, exist_ok=True)
        filename=(path_ROSTRO_REDIMENSIONADO+ '/' + cj + '/' + nj) 

        #Cambiar el tamaño resize
        height = 180
        width = 180
        dim = (width, height)
        redi_image = cv2.resize(imgBGR, dim, interpolation=cv2.INTER_LINEAR)

        cv2.imwrite(filename, redi_image) 


  print("TERMINO !!")
  print("Cantidad de imagenes ", cont_imagen)
```

```
def caso_opcion(opcion):
  if opcion == 1:
    print('Perros')
    path_MANUAL = '2.Resultados_Preprocesamiento/Perros/MANUALES' #1
    path_ROSTRO_REDIMENSIONADO='2.Resultados_Preprocesamiento/Perros/ROSTRO_REDIMENSIONADO' #2
    preprocesamiento_imagen(path_MANUAL,path_ROSTRO_REDIMENSIONADO)
  elif opcion == 2:
    print('Buscar Perro')
    path_MANUAL = '2.Resultados_Preprocesamiento/Buscar_perro/MANUALES' #1
    path_ROSTRO_REDIMENSIONADO='2.Resultados_Preprocesamiento/Buscar_perro/ROSTRO_REDIMENSIONADO' #2
    preprocesamiento_imagen(path_MANUAL,path_ROSTRO_REDIMENSIONADO)
  elif opcion == 3:
    print('Gatos')
    path_MANUAL = '2.Resultados_Preprocesamiento/Gatos/MANUALES' #1
    path_ROSTRO_REDIMENSIONADO='2.Resultados_Preprocesamiento/Gatos/ROSTRO_REDIMENSIONADO' #2
    preprocesamiento_imagen(path_MANUAL,path_ROSTRO_REDIMENSIONADO)
  elif opcion == 4:
    print('Buscar Gato')
    path_MANUAL = '2.Resultados_Preprocesamiento/Buscar_gato/MANUALES' #1
    path_ROSTRO_REDIMENSIONADO='2.Resultados_Preprocesamiento/Buscar_gato/ROSTRO_REDIMENSIONADO' #2
    preprocesamiento_imagen(path_MANUAL,path_ROSTRO_REDIMENSIONADO)
  elif opcion == 5:
    print('Gatos_QUERY')
    path_MANUAL = '2.Resultados_Preprocesamiento/Gatos_QUERY/MANUALES' #1
    path_ROSTRO_REDIMENSIONADO='2.Resultados_Preprocesamiento/Gatos_QUERY/ROSTRO_REDIMENSIONADO' #2
    preprocesamiento_imagen(path_MANUAL,path_ROSTRO_REDIMENSIONADO)
  elif opcion == 6:
    print('Perros_QUERY')
    path_MANUAL = '2.Resultados_Preprocesamiento/Perros_QUERY/MANUALES' #1
    path_ROSTRO_REDIMENSIONADO='2.Resultados_Preprocesamiento/Perros_QUERY/ROSTRO_REDIMENSIONADO' #2
    preprocesamiento_imagen(path_MANUAL,path_ROSTRO_REDIMENSIONADO)
  else:
    print('xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx')
    print('Error: La opción ingresada no esta dentro de las opciones, por favor vuelva a ingresar la opción')
    print('xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx')
    print('')
    print('')
    print('')
    seleccion_opcion()

seleccion_opcion()
```
