```
import dlib, cv2, os
import numpy as np
from skimage import io, color
import cv2 as cv
from matplotlib import pyplot as plt



# Estas carpetas estan en My Drive/0_Desarrollo_Tesis/2.Resultados_Preprocesamiento
path = 'ME_SIMILARIDAD' #1

array_imagenes =[] 
array_rutas=[]

cont_imagen=0
#For que recorre las carpetas(sujetos) del path
for aa in os.listdir(path):
    
    #For que recorre la imagen de cada carpeta(sujeto)
    for f in os.listdir(path + '/' + aa):

      cont_imagen=cont_imagen+1
      #aa Carpeta del sujeto 
      #f Nombre de la imagen del sujeto
      #Ruta de la imagen
      imgPath = path + '/' + aa + '/' + f

      array_rutas.append(imgPath)
      print(imgPath)
      #Abrir Imagen (Lee la imagen en formtado BGR)
      imgBGR = cv2.imread(imgPath)

      #Convertir una imagen a GRIS
      imgGRAY= color.rgb2gray(io.imread(imgPath))

      #Datos de la Imagen
      height, width = imgGRAY.shape
      #print(imgGRAY.shape)
      total=(height*width)

      #Sacar el Histograma
      plt.hist(imgGRAY.ravel(),256,[0,256]); plt.show()
      hist=[]
      hist = cv.calcHist([imgGRAY],[0],None,[256],[0,256])

      #Convertir el vector a un array
      array = np.array(hist)
      #Transponer el array
      array_tanspuesto=np.transpose(array)

      #Convertir de Matriz a Vector
      flat_arr = array_tanspuesto.ravel()
      #Convertir el vector a un array
      vector = np.array(flat_arr)
      #Se apila a el array
      array_imagenes.append(vector)
      
# Se tiene una matriz de todas las imagenes
matriz_imagenes=np.matrix(array_imagenes)

print(matriz_imagenes)  
print(len(matriz_imagenes))
print(matriz_imagenes.shape)
print(matriz_imagenes.size)
print(array_rutas)


print("Cantidad de imagenes PROCESADAS!! ", cont_imagen)
print("TERMINO !!")
```
